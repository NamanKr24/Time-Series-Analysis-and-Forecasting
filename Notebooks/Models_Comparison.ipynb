```python
import matplotlib.pyplot as plt

# Model names
models = ['ARIMA', 'LSTM', 'Transformer', 'Liquid ML', 'Hybrid Model']

# Placeholder RMSE values — replace these manually with your actual results
rmse_values = [19.8, 7.17, 12.5, 6.9, 6.53]

plt.figure(figsize=(8, 5))
bars = plt.bar(models, rmse_values, color=['skyblue', 'lightgreen', 'salmon', 'plum', 'brown'])

# Adding value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.02, f'{yval:.2f}', ha='center', va='bottom', fontsize=10)

plt.title('RMSE Comparison of Models')
plt.ylabel('RMSE')
plt.ylim(0, max(rmse_values) * 1.2)

plt.tight_layout()
plt.show()
```
